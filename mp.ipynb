```python
from khayyam import JalaliDate
```

```python
# Function to convert Hijri Shamsi date to Gregorian date
def convert_to_gregorian(hijri_date):
    year, month, day = map(int, hijri_date.split('/'))
    jalali_date = JalaliDate(year, month, day)
    gregorian_date = jalali_date.todate()
    return gregorian_date

# Apply the conversion function to the entire column
df['Gregorian Date'] = df['Gregorian Date'].apply(convert_to_gregorian)

df.head()
```

```python
# Convert Gregorian Date column to datetime format
df['Gregorian Date'] = pd.to_datetime(df['Gregorian Date'])

# Group the data by 'شماره پرونده'
grouped_data = df.groupby('شماره پرونده')

# Initialize an empty list to store the result
time_difference_data = []

# Iterate over each group
for group_name, group_df in grouped_data:
    # Calculate the earliest and latest datetime within each group
    earliest_datetime = group_df['Gregorian Date'].min() + pd.to_timedelta(group_df['Time'].astype(str).min())
    latest_datetime = group_df['Gregorian Date'].max() + pd.to_timedelta(group_df['Time'].astype(str).max())
    # Calculate the time difference
    time_difference = latest_datetime - earliest_datetime
    # Extract the value of 'گروه مشتري' for this group
    customer_group = group_df['گروه مشتري'].iloc[0]
    # Append the result to the time_difference_data list
    time_difference_data.append({'شماره پرونده': group_name, 'گروه مشتري': customer_group, 'Time Difference': time_difference})

# Create a DataFrame from the list
time_difference_df = pd.DataFrame(time_difference_data)

time_difference_df.head()
```

```python
# Group the data by 'گروه مشتري'
grouped_by_customer_group = time_difference_df.groupby('گروه مشتري')

# Calculate the average time difference for each 'گروه مشتري'
average_time_per_group = grouped_by_customer_group['Time Difference'].mean()

# Print the result
print(average_time_per_group)
```

```python
# Sort the average time differences for each 'گروه مشتري' in ascending order
sorted_average_time_per_group = average_time_per_group.sort_values(ascending=True)

# Print the result
print(sorted_average_time_per_group)
```

```python
# Filter the DataFrame to include only the specified values in 'وضعیت مرجوعی'
filtered_df = df[df['وضعیت مرجوعی'].isin(['مرجوع نشده', 'مرجوع شده'])]

# Group the filtered DataFrame by 'گروه مشتري' and 'وضعیت مرجوعی' and count the occurrences
counts_per_customer_group = filtered_df.groupby(['گروه مشتري', 'وضعیت مرجوعی']).size()

# Print the result
print(counts_per_customer_group)
```

```python
# Sort the counts of each value for each 'گروه مشتري' in ascending order
sorted_counts_per_customer_group = counts_per_customer_group.sort_values(ascending=True)

# Print the sorted result
print(sorted_counts_per_customer_group)
```

```python
# Filter the sorted counts to include only 'مرجوع شده' values
filtered_counts = sorted_counts_per_customer_group[sorted_counts_per_customer_group.index.get_level_values('وضعیت مرجوعی') == 'مرجوع شده']

# Print the filtered counts
print(filtered_counts)
```

```python
import pandas as pd

# Assuming df is your DataFrame containing the data
# Convert 'Gregorian Date' column to datetime format
df['Gregorian Date'] = pd.to_datetime(df['Gregorian Date'])

# Group the DataFrame by day of the week and month and count the occurrences
distribution_by_day_of_week = df.groupby(df['Gregorian Date'].dt.dayofweek).size()
distribution_by_month = df.groupby(df['Gregorian Date'].dt.month).size()

# Print the distributions
print("Distribution by day of the week:")
print(distribution_by_day_of_week)
print("\nDistribution by month:")
print(distribution_by_month)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest

# Plot histograms for distribution by day of the week and month
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Gregorian Date'].dt.dayofweek, bins=7, edgecolor='black', alpha=0.7)
plt.title('Distribution by Day of the Week')
plt.xlabel('Day of the Week (0 = Monday, 6 = Sunday)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['Gregorian Date'].dt.month, bins=12, edgecolor='black', alpha=0.7)
plt.title('Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Shapiro-Wilk test for normality
shapiro_day_of_week = shapiro(distribution_by_day_of_week)
shapiro_month = shapiro(distribution_by_month)

print("Shapiro-Wilk test for day of the week distribution (p-value):", shapiro_day_of_week[1])
print("Shapiro-Wilk test for month distribution (p-value):", shapiro_month[1])

# Kolmogorov-Smirnov test for normality
ks_day_of_week = kstest(distribution_by_day_of_week, 'norm')
ks_month = kstest(distribution_by_month, 'norm')

print("\nKolmogorov-Smirnov test for day of the week distribution (p-value):", ks_day_of_week.pvalue)
print("Kolmogorov-Smirnov test for month distribution (p-value):", ks_month.pvalue)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
graph = nx.DiGraph()

# Iterate over rows of the DataFrame
for index, row in df.iterrows():
    # Add nodes representing activities
    graph.add_node(row['فعاليت'])

    # Add edges representing connections between activities
    if index > 0:
        graph.add_edge(df.at[index - 1, 'فعاليت'], row['فعاليت'])

# Draw the graph
nx.draw(graph, with_labels=True, font_weight='bold')
plt.show()
```
